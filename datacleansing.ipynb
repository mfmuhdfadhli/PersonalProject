```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing required dataset from local directory and call it df
df = pd.read_csv("C:/Users/user/OneDrive/Desktop/Python Script/Soccer-Stats-Premier-League-2023-2024_Final_20240610.csv")
df.head()
```

```python
# a quick summary of our DataFrame
df.info()
```

```python
# Show rows where the 'name' column has NaN values
rows_with_nan_name = df[df['Name'].isna()]

# Display the rows with NaN in the 'name' column
print("Rows with NaN in 'Name' column:")
print(rows_with_nan_name)
```

```python
# Count the number of rows with NaN in the 'name' column
num_rows_with_nan_name = rows_with_nan_name.shape[0]

# Display the count of rows with NaN in the 'name' column
print(f"Total number of rows with NaN in 'name' column: {num_rows_with_nan_name}")
```

```python
# Remove rows where the 'name' column has NaN values
df_cleaned = df.dropna(subset=['Name'])

# Display the DataFrame after removing rows with NaN in the 'name' column
print("DataFrame after removing rows with NaN in 'Name' column:")
print(df_cleaned.head())
```

```python
# Print the shape of the cleaned DataFrame
print(f"Shape of the DataFrame after removing rows with NaN in 'name' column: {df_cleaned.shape}")

# Display the first few rows of the cleaned DataFrame
print("First few rows of the cleaned DataFrame:")
print(df_cleaned.head())
```

```python
# Display the DataFrame after removing rows with NaN in the 'name' column
print("DataFrame after removing rows with NaN in 'name' column:")
print(df_cleaned)
```

```python
# Creates a deep copy of the DataFrame df and assigns it to a new variable df_clean
df_clean = df.copy()

# Removes duplicate rows from the df_clean DataFrame based on the values in the column specified by the subset parameter, which in this case is the 'Name' column.
df_clean = df_clean.drop_duplicates(subset=['Name'])
```

```python
# Total number of rows that are considered duplicates
df_clean.duplicated(subset=['Name']).sum()
```

```python
# Display the DataFrame after removing duplicates in the 'name' column
print("DataFrame after removing duplicate values in 'name' column:")
print(df_cleaned)

# Print the shape of the cleaned DataFrame
print(f"Shape of the DataFrame after removing duplicates: {df_cleaned.shape}")

# Display the remaining rows to confirm
print("Remaining rows after removing duplicates:")
print(df_cleaned)
```

```python
# Show Null Values
df.isna().sum().sort_values(ascending=False)
```

```python
# Define the path where the cleaned CSV file will be saved
cleaned_file_path = 'cleaneddata.csv'

# Save the cleaned DataFrame to a new CSV file
df_cleaned.to_csv(cleaned_file_path, index=False)

print(f"Cleaned data has been saved to {cleaned_file_path}")
```
